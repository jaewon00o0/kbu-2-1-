```python
import requests
```

```python
res_naver = requests.get("https://www.naver.com")
res_google = requests.get("https://www.google.com")
```

```python
print(f"Server: {res_naver.headers.get('Server')}")
print(f"Date: {res_naver.headers.get('Date')}")
```

```python
print(f"Server: {res_google.headers.get('Server')}")
print(f"Date: {res_google.headers.get('Date')}")
```

```python
print("\n[네이버 전체 헤더]\n", res_naver.headers)
print("\n[구글 전체 헤더]\n", res_google.headers)
```
